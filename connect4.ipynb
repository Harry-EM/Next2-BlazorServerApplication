```python
class Connect4:
    def __init__(self):
        """Initialize a 6x7 board."""
        self.board = np.full((6, 7), ' ')
        self.current_player = '●'
        self.other_player = '○'
        self.horizontal_1 = 0
        self.horizontal_2 = 0
        self.vertical_1 = 0
        self.vertical_2 = 0
        self.diag_1 = 0
        self.diag_2 = 0

    def reset_board(self):
        """Reset the board to its original state. (Used for testing)"""
        self.board = np.full((6, 7), ' ')
        self.current_player = '●'
        self.other_player = '○'

    def print_board(self):
        """Display the current board."""
        for row in self.board:
            print("|".join(row))
            print("-" * 13)

    def get_moves(self):
      """Get all valid moves for self.board."""
      for i in range(7):
        if self.board[0, i] == ' ':
          for j in reversed(range(6)):
            if self.board[j, i] == ' ':
              yield (j, i)
              break

    def get_moves_ML(self, board):
      """Get all valid moves for board."""
      for i in range(7):
        if board[0, i] == ' ':
          for j in reversed(range(6)):
            if board[j, i] == ' ':
              yield (j, i)
              break


    def make_move(self, row, col, player):
        """Place the player's mark on the board if the move is valid."""
        if self.board[row, col] == ' ':
            self.board[row, col] = player
            return True
        return False

    def make_move_ML(self, row, col, player, board):
        """Place the player's mark on the board if the move is valid."""
        if board[row, col] == ' ':
            board[row, col] = player
            return board
        return board

    def is_winner(self, player):
        """Check if the given player has won."""
        rows, cols = 6, 7
        for row in range (rows):
          for col in range (cols):
            if col + 3 < cols and all(self.board[row, col + i] == player for i in range(4)):
              return True
            if row + 3 < rows and all(self.board[row + i, col] == player for i in range(4)):
              return True
            if row + 3 < rows and col + 3 < cols and all(self.board[row + i, col + i] == player for i in range(4)):
              return True
            if row + 3 < rows and col - 3 >= 0 and all(self.board[row + i, col - i] == player for i in range(4)):
              return True
        return False

    def is_winner_ML(self, player, board):
        """Check if the given player has won."""
        rows, cols = 6, 7
        for row in range (rows):
          for col in range (cols):
            if col + 3 < cols and all(board[row, col + i] == player for i in range(4)):
              return True
            if row + 3 < rows and all(board[row + i, col] == player for i in range(4)):
              return True
            if row + 3 < rows and col + 3 < cols and all(board[row + i, col + i] == player for i in range(4)):
              return True
            if row + 3 < rows and col - 3 >= 0 and all(board[row + i, col - i] == player for i in range(4)):
              return True
        return False

    def is_winner_testing(self, player):
        """Check if the given player has won."""
        rows, cols = 6, 7
        for row in range (rows):
          for col in range (cols):
            if col + 3 < cols and all(self.board[row, col + i] == player for i in range(4)):
              if player == '●':
                self.horizontal_1 += 1
              if player == '○':
                self.horizontal_2 += 1
              return True
            if row + 3 < rows and all(self.board[row + i, col] == player for i in range(4)):
              if player == '●':
                self.vertical_1 += 1
              if player == '○':
                self.vertical_2 += 1
              return True
            if row + 3 < rows and col + 3 < cols and all(self.board[row + i, col + i] == player for i in range(4)):
              if player == '●':
                self.diag_1 += 1
              if player == '○':
                self.diag_2 += 1
              return True
            if row + 3 < rows and col - 3 >= 0 and all(self.board[row + i, col - i] == player for i in range(4)):
              if player == '●':
                self.diag_1 += 1
              if player == '○':
                self.diag_2 += 1
              return True
        return False

    def is_3_in_a_row(self, player, board):
        """Check if the given player has 3 in a row."""
        rows, cols, num = 6, 7, 0
        for row in range (rows):
          for col in range (cols):
            if col + 2 < cols:
              if board[row, col] == player and board[row, col + 1] == player and board[row, col + 2] == player:
                num+=1
            if row + 2 < rows:
              if board[row, col] == player and board[row + 1, col] == player and board[row + 2, col] == player:
                num+=1
            if row + 2 < rows and col + 2 < cols:
              if board[row, col] == player and board[row + 1, col + 1] == player and board[row + 2, col + 2] == player:
                num+=1
            if row + 2 < rows and col - 2 > 0:
              if board[row, col] == player and board[row + 1, col - 1] == player and board[row + 2, col - 2] == player:
                num+=1
        return num

    def is_2_in_a_row(self, player, board):
        """Check if the given player has 2 in a row."""
        rows, cols, num = 6, 7, 0
        for row in range (rows):
          for col in range (cols):
            if col + 1 < cols:
               if board[row, col] == player and board[row, col + 1] == player:
                  num+=1
            if row + 1 < rows:
               if board[row, col] == player and board[row + 1, col] == player:
                 num+=1
            if row + 1 < rows and col + 1 < cols:
               if board[row, col] == player and board[row + 1, col + 1] == player:
                 num+=1
            if row + 1 < rows and col - 1 > 0:
               if board[row, col] == player and board[row + 1, col - 1] == player:
                 num+=1
        return num

    def is_full(self):
      """Check if self.board is full (draw)."""
      return ' ' not in self.board

    def is_full_ML(self, board):
      """Check if the board is full (draw)."""
      return ' ' not in board

    def getBottom(self, col, valid_moves):
      if self.board[0][col] == " ":
        if (5, col) in valid_moves:
          n=5
          print("5")
        elif (4, col) in valid_moves:
          n=4
          print("4")
        elif (3, col) in valid_moves:
          n=3
          print("3")
        elif (2, col) in valid_moves:
          n=2
          print("2")
        elif (1, col) in valid_moves:
          n=1
          print("1")
        else:
          n=0
          print("0")
      else:
        n=-1
        print("-1")
      return n

game = Connect4()
```

```python
def random_move(game):
  # Randomly place - can be done as it's own function or called from rule_based_move
  while True:
    moves = list(game.get_moves()) # Select all valid moves
    length=len(moves) # Get length of this list
    random = np.random.randint(length) # Randomly select one of these indexes
    move=moves[random] # Assign the selected index to that valid move
    game.make_move(move[0], move[1], game.current_player) # Place move
    break
```

```python
def rule_based_move(game):
    # start = time.time()
    move_complete=False
    board = game.board
    current_player = game.current_player
    other_player = game.other_player

    if game.is_full(): # If the board is full, don't try to place a counter
      move_complete=True

    moves = list(game.get_moves()) # Get all valid moves for this board

    # Take win if possible
    if not move_complete: # If the board isn't full, try to take the win
      for move in moves: # Try every valid move to see if it can win
        board[move[0], move[1]] = current_player
        if game.is_winner(current_player): # If that column placement is a winner
          board[move[0], move[1]] = " "
          game.make_move(move[0], move[1], current_player) # Make the move on the game board
          print(f"win at column:{move[1]} row:{move[0]}")
          move_complete=True
          break
        board[move[0], move[1]] = " " # Otherwise just remove counter

    # Block win from other player
      if not move_complete: # If no win can be taken, try to block the opponent's win
        for move in moves:
          board[move[0], move[1]] = other_player
          if game.is_winner(other_player): # If that column is a winner for the other player
            board[move[0], move[1]] = " "
            game.make_move(move[0], move[1], current_player) # Place your counter to avoid the loss
            print(f"Loss avoid at column:{move[1]} row:{move[0]}")
            move_complete=True
            break
          board[move[0], move[1]] = " " # Otherwise remove counter and try next valid column

    # Randomly place
    if not move_complete: # If there are no possible wins to take or block, place randomly
      random_move(game)
    print("Move made")
    # end = time.time()
    # print(f"Rule based time: {end1-start1}")
```

```python
#AI Agent Using Minimax with Alpha-Beta Pruning
class Minimax_Agent:
    def __init__(self, game, board, depth): # Input the game, board, and depth.
        """Initialize the AI agent with the game and players."""
        self.game = game
        self.board = board
        self.depth = depth # This allows the depth to be customised so that the ML algorithm is trained quicker. Lowering the depth does sacrifice accuracy however
        self.current_player = game.current_player
        self.other_player = game.other_player

    def minimax(self, depth, alpha, beta, is_maximizing, board):
        """Minimax algorithm with Alpha-Beta Pruning."""
        if self.game.is_winner_ML(self.current_player, board): # If the player has a winning move, the score will be really high
            return 1000-depth
        elif self.game.is_winner_ML(self.other_player, board): # If the other player has a winning move, the score will be really low
            return depth-1000
        elif self.game.is_full_ML(board): # If the board is filled, the score is 0 as it is better than losing but not as good as winning
            return 0
        elif depth == 0: # If the maximum depth is reached, find the return the score for that board
            return self.score_placement(board)


        if is_maximizing: # If it is maximising the score
            max_eval = float('-inf') # Set initial value to lowest possible
            for (row, col) in self.game.get_moves_ML(board): # For each valid move in the current board state
                board[row, col] = self.current_player # Place the counter in each valid column
                eval = self.minimax(depth - 1, alpha, beta, False, board) # Use the minimax function on the lower down branch
                board[row, col] = ' ' # set it back to blank
                max_eval = max(max_eval, eval) # find if the eval (score) is higher than the maximum eval (score)
                alpha = max(alpha, eval) # Find if the current eval is higher than the alpha value and if it is set it
                if beta <= alpha: # If beta is less than or equal to alpha then you won't need to check this branch any more since the algorithm will always choose beta
                    break  # Beta cut-off
            return max_eval
        else: # Otherwise it is minimising the score
          min_eval = float('inf') # Set initial value to highest possible
          for (row, col) in self.game.get_moves_ML(board):
              board[row, col] = self.other_player
              eval = self.minimax(depth - 1, alpha, beta, True, board) # Use minimax on lower down branch
              board[row, col] = ' '
              min_eval = min(min_eval, eval) # Same as maximising but with beta values and the opposite player minimising
              beta = min(beta, eval)
              if beta <= alpha:
                  break  # Alpha cut-off
          return min_eval

    def score_placement(self, board):
        """Evaluate the placement of a move."""
        score = 0
        score += self.game.is_3_in_a_row(self.current_player, board) * 10
        score -= self.game.is_3_in_a_row(self.other_player, board) * 15
        score += self.game.is_2_in_a_row(self.current_player, board) * 3
        score -= self.game.is_2_in_a_row(self.other_player, board) * 3

        for i in range(6):
          if board[i, 3] == self.current_player:
            score += 7
          elif board[i, 3] == self.other_player:
            score -= 3
          if board[i, 2] == self.current_player:
            score += 5
          elif board[i, 2] == self.other_player:
            score -= 1
          if board[i, 4] == self.current_player:
            score += 5
          elif board[i, 4] == self.other_player:
            score -= 1
          if board[i, 1] == self.current_player:
            score += 2
          if board[i, 5] == self.current_player:
            score += 2
        return score

    def minimax_move(self):
        """Get the best move for AI using Minimax."""
        best_val = float('-inf')
        move = None
        new_copy = copy.deepcopy(self.board) # https://www.geeksforgeeks.org/array-copying-in-python/
        # I deep copy here to ensure that it doesn't affect the board at all
        for (row, col) in self.game.get_moves_ML(new_copy): # Goes through every valid move
          new_board = self.game.make_move_ML(row, col, self.current_player, copy.deepcopy(new_copy)) # Make each valid move to the board
          move_val = self.minimax(self.depth, float('-inf'), float('inf'), False, new_board) # Minimax algorithm on that board
          new_board[row, col] = ' ' # Reset the board back to the original state
          if move_val > best_val: # If the value of any of the moves in that originl column placement have a greater value than the best value
              best_val = move_val # The best value becomes that value
              move = (row, col)
        #print(move, best_val)
        return move

    def make_move(self, current_player): # Makes the move - only used if minimax algorithm is called and not for training the data
      #start1 = time.time()
      self.current_player = current_player
      self.other_player = '○' if current_player == '●' else '●'
      row, col = self.minimax_move()
      self.game.make_move(row, col, self.current_player)
      #end1 = time.time()
      #print(f"Minimax time: {end1-start1}")
```

```python
data = pd.read_csv("/content/drive/MyDrive/connect-4.data", header=None)
# start = time.time()
game = Connect4()
board_cols = [f'pos_{i}' for i in range(42)]
data.columns = board_cols + ['outcome']
print(data[board_cols].head())
print("Dataset preview:")
print(data.head())

token_mapping = {'x': 1, 'o': -1, 'b': 0}
for col in board_cols:
    data[col] = data[col].map(token_mapping)

def infer_move(board_tokens): # Used to find the best move for each board data point
    board_tokens_formatted = np.full((6, 7), ' ') # Create blank board
    for row in range(6):
      for col in range(7): # Converts flat board to 2D array board
        idx = col * 6 + row
        token = board_tokens[idx]
        if token == 1:
          space = game.current_player
        elif token == -1:
          space = game.other_player
        else:
          space = ' '
        board_tokens_formatted[5-row, col] = space

    agent = Minimax_Agent(game, board_tokens_formatted, 1)
    row, col = agent.minimax_move() # My minimax algoritm is used with a depth of 1 to determine the best move
    # I have used a depth of 1 to ensure that it is effective yet doesn't take too long to train.
    idx = col
    return idx

def encode_board(board_tokens):
    return [token for token in board_tokens]

X_data = []
y_data = []

for idx, row in data.iterrows(): # Creates set of every board in dataset
    board_tokens = row[board_cols].tolist() # Converts values (e.g b, b, x, o... into a list [b, b, x, o...])
    move = infer_move(board_tokens) # Finds best move for data point
    if move is not None: # If there is a valid move
        X_data.append(encode_board(board_tokens)) # Adds board to X_data
        y_data.append(move) # Adds best move to y_data

X_data = np.array(X_data) # Creates 2D array of all valid boards
y_data = np.array(y_data) # Creates array of all best moves

print(f"Total training samples: {X_data.shape[0]}")
print("Example board:", X_data[0])
print("Inferred move for that board:", y_data[0])

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
print(X_train , "X data", len(X_train))
print(y_train, "Y data", len(y_train))
model = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=10000, random_state=42)

model.fit(X_train, y_train)

train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)
print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
# end = time.time()
# print(f"Training time: {end-start}")
```
